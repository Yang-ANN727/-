```python
# 绘制 treatment 与 outcome 的平均趋势示例
import seaborn as sns
import matplotlib.pyplot as plt
df_group = df.groupby(['year','treatment'])['outcome'].mean().reset_index()
sns.lineplot(data=df_group, x='year', y='outcome', hue='treatment')
plt.title('Average outcome by treatment over time')
plt.show()
```
